<b>Question 6</b><br>

Let $e_1$ and $e_2$ be the standard unit vectors in $\mathbb{R}^2$. Prove that $e_1$, $e_1 + e_2$ are linearly independent

<b> Solution</b><br>

Lets take scalars as $k_1, k_2$ and the linear combination as :

$k_1(e_1) + k_2(e_1 + e_2) = 0$

$\Rightarrow k_1e_1 + k_2e_1 + k_2e_2 = 0$

$\Rightarrow (k_1 + k_2)e_1 + k_2e_2 = 0$

Since, $e_1, e_2$ are standard unit vectors(or linearly independent as proven earlier), therefore, scalars must be 0 ie. $k_1 + k_2 = 0$ and $k_2 = 0$ 

$\Rightarrow k_1 = 2$ and $k_2 = 0$

Therefore, $e_1$ and $e_1 + e_2$ linearly independent

If $k_1 = 1$, then 

$k_2 = -1$ <br>
$k_3 = 1$<br>
$k_4 = -1$

This gives us scalar values of $eq(1)$ are not zeroes.

<b>Question 9</b><br>

Let $u$, $v$, $w$ be linearly independent vectors in $\mathbb{R}^n$. Let the vector $x$ be in $\mathbb{R}^n$ such that $x = k_1 u + k_2 v + k_3 w$, where the $k's$ are real scalars. Prove that this representation of the vector $x$ is unique.

<b>Solution</b> <br>

Lets take another real scalars $c's$ as $c_1$, $c_2$, and $c_3$ such that

$x = c_1u + c_2v + c_3w$

We have, <br>
$x = k_1 u + k_2 v + k_3 w$

Therefore, <br>
$c_1u + c_2v + c_3w = k_1 u + k_2 v + k_3$

$\Rightarrow c_1 - k_1u + c_2v - k_2v + c_3w - k_3w = 0$

$\Rightarrow (c_1 - k_1)u + (c_2 - k_2)v + (c_3 - k_3)w = 0$

As, $u, v, w$ are linearly independent, <br>
$c_1 - k_1 = 0\qquad \Rightarrow c_1 = k_1$<br>
$c_2 - k_2 = 0\qquad \Rightarrow c_2 = k_2$<br>
$c_3 - k_3 = 0\qquad \Rightarrow c_3 = k_3$

Hence, <b>unique</b>

<b>Question 10</b><br>

Let $S = \{v_1, v_2, v_3, ..., v_n\}$ be a set of linear independent vectors in $\mathbb{R}^n$. <br>Prove that  $T = \{c_1v_1, c_2v_2, c_3v_3, ..., c_nv_n\}$, where the $c's$ are real non-zero scalars, is also a set of linear independent vectors.

<b> Solution</b>

Lets take scalars $k_1, k_2, \dots k_n$ and linear combination of $T$ set of vectors as: 

$$k_1(c_1v_1) + k_2(c_2v_2) + k_3(c_3v_3) + \cdots k_n(c_nv_n) = 0$$

We know that,<br>
$k(u\cdot v) = (ku)\cdot v$

Threfore, above equation can be written as :

$(k_1c_1)v_1 + (k_2c_2)v_2 + (k_3c_3)v_3 + \dots + (k_nc_n)v_n = 0$

We are given that, set of vectors $S = \{v_1, v_2, v_3, ..., v_n\}$ are linearly independent. Therefore, their scalars must equate to zero.

$\Rightarrow k_1c_1 = k_2c_2 = \dots =k_nc_n = 0$

$\Rightarrow k_1 = k_2 = k_3 = 0$<br>
Therefore, T set of vectors must also be linearly independent
