<b>Question 3</b>

Find the eigenvalues and eigenvectors of $A = \begin{pmatrix}
3&1\\
1&3\\
\end{pmatrix}$. Plot the eigenspaces $E_{\lambda}$

<b>Solution</b> 

To find eigenvalues, we need to solve $det(A - \lambda I) = 0$

$\Rightarrow det\left(\begin{pmatrix}
3&1\\
1&3\\
\end{pmatrix} - \lambda \begin{pmatrix}
1&0\\
0&1\\
\end{pmatrix}\right) = 0$

$\Rightarrow det\begin{pmatrix}
3-\lambda&1\\
1&3-\lambda\\
\end{pmatrix} = 0$

$\Rightarrow (3-\lambda)^2 - 1 = 0$<br>
$\Rightarrow (3-\lambda)^2 = 1$<br>
$\Rightarrow 3 - \lambda = -1$ or $3 -\lambda =1$<br>


$\Rightarrow \lambda = 4$ or $\lambda =2$

If $\lambda = 4$, then $(A - \lambda I)u = 0$

$\Rightarrow \left(\begin{pmatrix}
3&1\\
1&3\\
\end{pmatrix} -4\begin{pmatrix}
1&0\\
0&1\\
\end{pmatrix}\right)u = 0$

$\Rightarrow \begin{pmatrix}
-1&1\\
-1&1\\
\end{pmatrix} \cdot \begin{pmatrix}
x\\
y\\
\end{pmatrix} = \begin{pmatrix}
0\\
0\\
\end{pmatrix}$

$\Rightarrow -x + y = 0 \Rightarrow y= x$<br>
$\Rightarrow x - y = 0 \Rightarrow x = y$<br>

Therefore, eigenvector for $4\lambda = 4$, is $u = \begin{pmatrix}
1\\
1\\
\end{pmatrix}$


If $\lambda = 2$, then $(A - \lambda I)u = 0$

$\Rightarrow \left(\begin{pmatrix}
3&1\\
1&3\\
\end{pmatrix} -2\begin{pmatrix}
1&0\\
0&1\\
\end{pmatrix}\right)u = 0$

$\Rightarrow \begin{pmatrix}
1&1\\
1&1\\
\end{pmatrix} \cdot \begin{pmatrix}
x\\
y\\
\end{pmatrix} = \begin{pmatrix}
0\\
0\\
\end{pmatrix}$

$\Rightarrow x + y = 0$<br>
$\Rightarrow x + y = 0$<br>

$\Rightarrow x = -y$

Therefore, eigenvector for $4\lambda = 4$, is $u = \begin{pmatrix}
-1\\
1\\
\end{pmatrix}$

```python
import numpy as np
from scipy.linalg import eig
import matplotlib.pyplot as plt

# Compute the eigenvalues and eigenvectors of the matrix
A = np.array([[3, 1],[1, 3]])
eigenvalues, eigenvectors = eig(A)
v1, v2 = eigenvectors[:, 0], eigenvectors[:, 1]


plt.figure(figsize=(10,8))

# Plot a line based on the first eigenvector
x1 = np.linspace(-5, 5, 100)
y1 = v1[1] / v1[0] * x1
plt.plot(x1, y1, 'r', label='Eigenspace 1')

# Plot a line based on the second eigenvector
x2 = np.linspace(-5, 5, 100)
y2 = v2[1] / v2[0] * x2
plt.plot(x2, y2, 'g', label='Eigenspace 2')

# Set plot limits and add grid lines
plt.xlim(-5, 5)
plt.ylim(-5, 5)
plt.grid()

# Plot x and y axes crossing the origin at (0, 0)
plt.axhline(0, color='k', linewidth=1)
plt.axvline(0, color='k', linewidth=1)

for x_val in range(-5, 6):
    plt.plot(x_val, 0, 'k+', markersize=5)
    plt.text(x_val, -0.5, f'{x_val}')

for y_val in range(-5, 6):
    plt.plot(0, y_val, 'k+', markersize=5)
    plt.text(-0.5, y_val, f'{y_val}')

plt.legend(loc="upper right")
plt.show()
```

Let $A$ be a square matrix and $B = rA$ where $r$ is a real number. Prove that if $\lambda$ is the eigenvalue of matrix $A$ then the eigenvalue of $B$ is $r\lambda$

<b> Solution</b> 

If $\lambda$ is the eigenvalue and $u$ is the eigenvector of $A$<br>
Then, <br>
$Au = \lambda u$<br>
and, as $B = rA$

We have, <br>
$Bu = (rA)u$<br>
$\quad = r(Au)$<br>
$\quad = r(\lambda u)$<br>
$\quad = (r\lambda)u$<br>

$\Rightarrow Bu = (r\lambda)u$<br>

Therefore, $r\lambda$ is the eigenvalue of B
