<b>Question 1</b> 

Solve the following linear system $Ax = b$ by using LU factorization for: 

(a) $A = \begin{pmatrix}
1&2&2\\
3&-3&-2\\
4&-1&-5\\
\end{pmatrix}, b = \begin{pmatrix}
5\\
0\\
-10\\
\end{pmatrix}\qquad$(b) $A = \begin{pmatrix}
1&5&6\\
2&11&19\\
3&19&47\\
\end{pmatrix}, b = \begin{pmatrix}
1\\
-4\\
-22\\
\end{pmatrix}$

(c) $A = \begin{pmatrix}
2&1&3\\
4&-1&0\\
10&12&34\\
\end{pmatrix}, b = \begin{pmatrix}
5\\
-11\\
84\\
\end{pmatrix}\qquad$(d) $A = \begin{pmatrix}
1&2&3\\
3&-9&-12\\
-1&8&18\\
\end{pmatrix}, b = \begin{pmatrix}
-5\\
9\\
-4\\
\end{pmatrix}$

<b> Solution</b>

<b>LU Factorization Procedure</b>

Given a matrix $A$, we can form an upper triangular matrix $U$ by using two row operations:<br>
- Multiplying a row by a non-zero constant
- Adding a multiple of one row to another

Swapping rows is <i>not allowed</i> 

<b>Solution (a)</b>

$A = \begin{pmatrix}
1&2&2\\
3&-3&-2\\
4&-1&-5\\
\end{pmatrix}$

Now, to get $U$ - upper triangular matrix

Step 1: $R_2 = R_{2} - 3R_{1}$

$\begin{pmatrix}
1&2&2\\
0&-9&-8\\
4&-1&-5\\
\end{pmatrix}$

Step 2: $R_3 = R_{3} - 4R_{1}$

$\begin{pmatrix}
1&2&2\\
0&-9&-8\\
0&-9&-13\\
\end{pmatrix}$

Step 3: $R_3 = R_{3} - R_{2}$

$\begin{pmatrix}
1&2&2\\
0&-9&-8\\
0&0&-5\\
\end{pmatrix} = U$ This is the upper triangular matrix, $U$

Now, using negative multipliers from the step1 , step2, and step3, and with $I$ $3 \times 3$ matrix<br>
We get, <br>
$L = \begin{pmatrix}
1&0&0\\
3&1&0\\
4&1&1\\
\end{pmatrix}$

Next, we can solve <b>$Ly = b$</b> to get y:

$\begin{pmatrix}
1&0&0\\
3&1&0\\
4&1&1\\
\end{pmatrix} \cdot \begin{pmatrix}
y_{1}\\
y_{2}\\
y_{3}\\
\end{pmatrix} = \begin{pmatrix}
5\\
0\\
-10\\
\end{pmatrix}$

Here, $y_{1} = 5$<br>
$\Rightarrow 3y_{1} + y_{2} =0$<br>
$\Rightarrow y_{2} = -3y_{1} = -15$<br>

$\Rightarrow y_{3} = 4(5) + (-15) + y_{3} = 0$<br>
$\Rightarrow y_{3} = -15$<br>

Next, we need to solve $Ux = y$ to get our solution<br>
We have now, <br>
$\begin{pmatrix}
1&2&2\\
0&-9&-8\\
0&0&-5\\
\end{pmatrix} \cdot \begin{pmatrix}
x_{1}\\
x_{2}\\
x_{3}\\
\end{pmatrix} = \begin{pmatrix}
5\\
-15\\
-15\\
\end{pmatrix}$

Here, $-5x_{3} = -15$<br>
$\Rightarrow x_{3} = 3$

Next, <br>
$-9x_{2} -8x_{3} = -15$<br>
$\Rightarrow -9x_{2} -8(3) = -15$<br>
$\Rightarrow -9x_{2} = -15 + 24$<br>
$\Rightarrow x_{2} = -1$<br>

Next, <br>
$x_{1} +2x_{2} + 2x_{3} = 5$<br>
$\Rightarrow x_{1} +2(-1) +2(3) = 5$<br>
$\Rightarrow x_{1} -2 + 6 = 5$<br>
$\Rightarrow x_{1} = 1$<br>

Hence, <b>$x_{1} =1, x_{2} = -1$ and $x_{3} = 3$</b>

<b>Question 2</b>

Solve the following system $Ax = b$ by using LU factorization:

$A = \begin{pmatrix}
1&2&3&4\\
17&22&27&8\\
77&44&47&-494\\
-10&1&7&63\\
\end{pmatrix}, b = \begin{pmatrix}
-10\\
22\\
2106\\
-243\\
\end{pmatrix}$

<b>Solution</b>

Step 1 : $R_{2} = R_{2} - 17R_{1}$

$\begin{pmatrix}
1&2&3&4\\
0&-12&-24&60\\
77&44&47&-494\\
-10&1&7&63\\
\end{pmatrix}$

Step 2: $R_{2} = \frac{1}{-12}R_{2}$

$\begin{pmatrix}
1&2&3&4\\
0&1&2&5\\
77&44&47&-494\\
-10&1&7&63\\
\end{pmatrix}$

Step 3: $R_{3} = R_{3} - 77R_{1}$

$\begin{pmatrix}
1&2&3&4\\
0&1&2&5\\
0&-110&-184&-802\\
-10&1&7&63\\
\end{pmatrix}$

Step 4: $R_{3} = R_{3} + 110R_{2}$

$\begin{pmatrix}
1&2&3&4\\
0&1&2&5\\
0&0&36&-252\\
-10&1&7&63\\
\end{pmatrix}$

Step 5: $R_{4} = R_{4} + 10R_{1}$

$\begin{pmatrix}
1&2&3&4\\
0&1&2&5\\
0&0&36&-252\\
0&21&37&103\\
\end{pmatrix}$

Step 6: $R_{4} = R_{4} - 21R_{2}$

$\begin{pmatrix}
1&2&3&4\\
0&1&2&5\\
0&0&36&-252\\

\end{pmatrix}$

Step 7: $R_{3} = \frac{1}{36}R_{3}$

$\begin{pmatrix}
1&2&3&4\\
0&1&2&5\\
0&0&1&-7\\
0&0&-5&-2\\
\end{pmatrix}$

Step 8: $R_{4} = R_{4} - 5R_{3}$

$\begin{pmatrix}
1&2&3&4\\
0&1&2&5\\
0&0&1&-7\\
0&0&0&-37\\
\end{pmatrix} = U$

Collecting negative multipliers from $step1, \dots step8$ we can get Lower triangular matrix, $L$ as: 

$L = \begin{pmatrix}
1&0&0&0\\
17&-12&0&0\\
77&-110&36&0\\
-10&21&-5&1\\
\end{pmatrix}$

Next, solving $Ly = b$ to get y<br>
$\begin{pmatrix}
1&0&0&0\\
17&-12&0&0\\
77&-110&36&0\\
-10&21&-5&1\\
\end{pmatrix} \cdot \begin{pmatrix}
y_1\\
y_2\\
y_3\\
y_4\\
\end{pmatrix} = \begin{pmatrix}
-10\\
22\\
2106\\
-243\\
\end{pmatrix}$

$\Rightarrow y_{1} = -10$

Next,<br>
$17y_{1} - 12y_{2} = 22$<br>
$\Rightarrow 17(-10) - 12y_{2} = 22$<br>
$\Rightarrow -170 - 12y_{2} = 22$

$\Rightarrow y_{2} = \frac{192}{-12} = -16$

Then<br>
$77y_{1} -110y_{2} + 36y_{3} = 2106$<br>
$\Rightarrow 77(-10) -110(-16) + 36y_{3} = 2106$<br>
$\Rightarrow -770 + 1760 + 36y_{3} = 2106$<br>
$\Rightarrow 36y_{3} = 2106 - 990$

$\Rightarrow y_{3} = \frac{1116}{36} = 31$

Then, <br>
$-10y_{1} + 21y_{2} - 5y_{3} + y_{4} = -243$<br>
$-10(-10) + 21(-16) -5(31) + y_{4} = -243$<br>
$\Rightarrow 100 -336 - 155 + y_{4} = -243$

$\Rightarrow y_{4} = -243 + 391 = 148$


Now, solving $Ux = y$ to get the solution

$\begin{pmatrix}
1&2&3&4\\
0&1&2&5\\
0&0&1&-7\\
0&0&0&-37\\
\end{pmatrix} \cdot \begin{pmatrix}
x_{1}\\
x_{2}\\
x_{3}\\
x_{4}\\
\end{pmatrix} = \begin{pmatrix}
-10\\
-16\\
31\\
148\\
\end{pmatrix}$

Solving this, we get <b>our solution as $x_{1} = 1, x_{2} = -2, x_{3} = 3$ and $x_{4} = -4$

<b> Question 3</b>

Find the determinant of the matrix $A$ given in question 2.

<b> Solution</b> 

We can write given matrix A as $A = LU$ representing :

$A =  \begin{pmatrix}
1&0&0&0\\
17&-12&0&0\\
77&-110&36&0\\
-10&21&-5&1\\
\end{pmatrix} \cdot \begin{pmatrix}
1&2&3&4\\
0&1&2&5\\
0&0&1&-7\\
0&0&0&-37\\
\end{pmatrix}$

Then, $det(A)$ is given by $det(A) = det(LU)$

$\Rightarrow det(A) = det(L) \times det(U)$

$det(A) = \left[1 \times (-12) \times 36 \times1)\right] \times \left[1 \times 1 \times 1 \times - 37 \right]$

$det(A) = 15984$
