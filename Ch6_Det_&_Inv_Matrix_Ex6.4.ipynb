<b>Question 1</b> 

Solve the following linear system $Ax = b$ by using LU factorization for: 

(a) $A = \begin{pmatrix}
1&2&2\\
3&-3&-2\\
4&-1&-5\\
\end{pmatrix}, b = \begin{pmatrix}
5\\
0\\
-10\\
\end{pmatrix}\qquad$(b) $A = \begin{pmatrix}
1&5&6\\
2&11&19\\
3&19&47\\
\end{pmatrix}, b = \begin{pmatrix}
1\\
-4\\
-22\\
\end{pmatrix}$

(c) $A = \begin{pmatrix}
2&1&3\\
4&-1&0\\
10&12&34\\
\end{pmatrix}, b = \begin{pmatrix}
5\\
-11\\
84\\
\end{pmatrix}\qquad$(d) $A = \begin{pmatrix}
1&2&3\\
3&-9&-12\\
-1&8&18\\
\end{pmatrix}, b = \begin{pmatrix}
-5\\
9\\
-4\\
\end{pmatrix}$

<b> Solution</b>

<b>LU Factorization Procedure</b>

Given a matrix $A$, we can form an upper triangular matrix $U$ by using two row operations:<br>
- Multiplying a row by a non-zero constant
- Adding a multiple of one row to another

Swapping rows is <i>not allowed</i> 

<b>Solution (a)</b>

$A = \begin{pmatrix}
1&2&2\\
3&-3&-2\\
4&-1&-5\\
\end{pmatrix}$

Now, to get $U$ - upper triangular matrix

Step 1: $R_2 = R_{2} - 3R_{1}$

$\begin{pmatrix}
1&2&2\\
0&-9&-8\\
4&-1&-5\\
\end{pmatrix}$

Step 2: $R_3 = R_{3} - 4R_{1}$

$\begin{pmatrix}
1&2&2\\
0&-9&-8\\
0&-9&-13\\
\end{pmatrix}$

Step 3: $R_3 = R_{3} - R_{2}$

$\begin{pmatrix}
1&2&2\\
0&-9&-8\\
0&0&-5\\
\end{pmatrix} = U$ This is the upper triangular matrix, $U$

Now, using negative multipliers from the step1 , step2, and step3, and with $I$ $3 \times 3$ matrix<br>
We get, <br>
$L = \begin{pmatrix}
1&0&0\\
3&1&0\\
4&1&1\\
\end{pmatrix}$

Next, we can solve <b>$Ly = b$</b> to get y:

$\begin{pmatrix}
1&0&0\\
3&1&0\\
4&1&1\\
\end{pmatrix} \cdot \begin{pmatrix}
y_{1}\\
y_{2}\\
y_{3}\\
\end{pmatrix} = \begin{pmatrix}
5\\
0\\
-10\\
\end{pmatrix}$

Here, $y_{1} = 5$<br>
$\Rightarrow 3y_{1} + y_{2} =0$<br>
$\Rightarrow y_{2} = -3y_{1} = -15$<br>

$\Rightarrow y_{3} = 4(5) + (-15) + y_{3} = 0$<br>
$\Rightarrow y_{3} = -15$<br>

Next, we need to solve $Ux = y$ to get our solution<br>
We have now, <br>
$\begin{pmatrix}
1&2&2\\
0&-9&-8\\
0&0&-5\\
\end{pmatrix} \cdot \begin{pmatrix}
x_{1}\\
x_{2}\\
x_{3}\\
\end{pmatrix} = \begin{pmatrix}
5\\
-15\\
-15\\
\end{pmatrix}$

Here, $-5x_{3} = -15$<br>
$\Rightarrow x_{3} = 3$

Next, <br>
$-9x_{2} -8x_{3} = -15$<br>
$\Rightarrow -9x_{2} -8(3) = -15$<br>
$\Rightarrow -9x_{2} = -15 + 24$<br>
$\Rightarrow x_{2} = -1$<br>

Next, <br>
$x_{1} +2x_{2} + 2x_{3} = 5$<br>
$\Rightarrow x_{1} +2(-1) +2(3) = 5$<br>
$\Rightarrow x_{1} -2 + 6 = 5$<br>
$\Rightarrow x_{1} = 1$<br>

Hence, <b>$x_{1} =1, x_{2} = -1$ and $x_{3} = 3$</b>

<b>Solution(b)</b>

$A = \begin{pmatrix}
1&5&6\\
2&11&19\\
3&19&47\\
\end{pmatrix}$

Now, to get $U$ - upper triangular matrix

Step 1: $R_2 = R_{2} - 2R_{1}$

$\begin{pmatrix}
1&5&6\\
0&1&7\\
3&19&47\\
\end{pmatrix}$

Step 2: $R_3 = R_{3} - 3R_{1}$

$\begin{pmatrix}
1&5&6\\
0&1&7\\
0&4&29\\
\end{pmatrix}$

Step 3: $R_3 = R_{3} - 4R_{2}$

$\begin{pmatrix}
1&5&6\\
0&1&7\\
0&0&3\\
\end{pmatrix} = U$ This is the upper triangular matrix, $U$

Now, using negative multipliers from the step1 , step2, and step3, and with $3 \times 3$ matrix $I$<br>
We get, <br>
$L = \begin{pmatrix}
1&0&0\\
2&1&0\\
3&4&1\\
\end{pmatrix}$

Next, we can solve <b>$Ly = b$</b> to get y:

$\begin{pmatrix}
1&0&0\\
2&1&0\\
3&4&1\\
\end{pmatrix} \cdot \begin{pmatrix}
y_{1}\\
y_{2}\\
y_{3}\\
\end{pmatrix} = \begin{pmatrix}
1\\
-4\\
-22\\
\end{pmatrix}$

$\Rightarrow y_{1} = 1$<br>
Then, <br>
$2y_{1} + y_{2} = -4 \qquad \Rightarrow y_{2} = -4 -2 = -6$<br>

and, $3y_{1} + 4y_{2} + y_{3} = -22$<br>
$\Rightarrow 3(1) +4(-6) + y_{3} = -22$<br>
$\Rightarrow 3 - 24 + y_{3} = -22$<br>

$\Rightarrow y_{3} = -22 + 21 = -1$

Next, we need to solve $Ux = y$ to get our solution<br>
We have now, <br>
$\begin{pmatrix}
1&5&6\\
0&1&7\\
0&0&3\\
\end{pmatrix} \cdot \begin{pmatrix}
x_{1}\\
x_{2}\\
x_{3}\\
\end{pmatrix} = \begin{pmatrix}
1\\
-6\\
-1\\
\end{pmatrix}$

$\Rightarrow x_{3} = -1$
Then,<br>
$x_{2} + 7x_{3} = -6$<br>
$\Rightarrow x_{2} =-6 -(-7) = 1$<br>
and, <br>
$x_{1} + 5(1) + 6(-1) = 1$<br>
$\Rightarrow x_{1} + 5 - 6 = 1$<br>
$\Rightarrow x_{1} = 2$

Hence, $x_{1} = 2, x_{2} = 1$ and $x_{1} = -1$ is our solution


<b>Solution(c)</b>

$A = \begin{pmatrix}
2&1&3\\
4&-1&0\\
10&12&34\\
\end{pmatrix}$

Now, to get $U$ - upper triangular matrix

Step 1: $R_2 = R_{2} - 2R_{1}$

$\begin{pmatrix}
2&1&3\\
0&-3&-6\\
10&12&34\\
\end{pmatrix}$

Step 2: $R_3 = R_{3} - 5R_{1}$

$\begin{pmatrix}
2&1&3\\
0&-3&-6\\
0&7&19\\
\end{pmatrix}$

Step 3: $R_2 = R_{2} - \frac{1}{-3}R_{2}$

$\begin{pmatrix}
2&1&3\\
0&1&2\\
0&7&19\\
\end{pmatrix}$

Step 4: $R_{3} = R_{3} - 7R_{3}$

$\begin{pmatrix}
2&1&3\\
0&1&2\\
0&0&5\\
\end{pmatrix}= U$ This is the upper triangular matrix, $U$

Now, using negative multipliers from the step1 , step2, step3, and step4, and with $3 \times 3$ matrix $I$<br>
We get, <br>
$L = \begin{pmatrix}
1&0&0\\
2&-3&0\\
5&7&1\\
\end{pmatrix}$

Next, we can solve <b>$Ly = b$</b> to get y:

$\begin{pmatrix}
1&0&0\\
2&-3&0\\
5&7&1\\
\end{pmatrix} \cdot \begin{pmatrix}
y_{1}\\
y_{2}\\
y_{3}\\
\end{pmatrix} = \begin{pmatrix}
5\\
-11\\
84\\
\end{pmatrix}$

$\Rightarrow y_{1} = 5$<br>
Next,<br>
$2y_{1} -3y_{2} = -11$<br>
$-3y_{2} = -11 - 10$<br>
$\Rightarrow y_{2} = 7$

$5y_{1} + 7y_{2} + y_{3} = 84$<br>
$5(5) + 7(7) + y_{3} = 84$<br>
$\Rightarrow y_{3} = 84 - 74 = 10$

Next, we have to solve $Ux = y$ to get our solution

$\begin{pmatrix}
2&1&3\\
0&1&2\\
0&0&5\\
\end{pmatrix} \cdot \begin{pmatrix}
x_{1}\\
x_{2}\\
x_{3}\\
\end{pmatrix} = \begin{pmatrix}
5\\
7\\
10\\
\end{pmatrix}$

Solving this, we get $x_{1} = -2,, x_{2} = 3$, and $x_{3} = 2$
