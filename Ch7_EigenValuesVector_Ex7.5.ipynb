<b> Solution (a) </b> 

Here, $A^TA = \begin{pmatrix}
1&0\\

\end{pmatrix} \begin{pmatrix}
1&0\\

\end{pmatrix}$

$\Rightarrow A^TA = \begin{pmatrix}

\end{pmatrix}$

Eigenvalues of $A^TA: det(A - \lambda I) = 0$

$\Rightarrow \begin{pmatrix}
1- \lambda&0\\
0 & 4 - \lambda\\
\end{pmatrix} = 0$

$\Rightarrow (1- \lambda)(4 - \lambda) = 0$

$\Rightarrow \lambda_{1} =1$ and $\lambda_{2} = 4$

We know that, <br>
If $A$ is $m$ by $n$ matrix and $\lambda_{1}, \lambda_{2},...,\lambda_{n}$ be the eigenvalues of $A^TA$,, then singular values denoted by $\sigma_{1}, \sigma_{2},...\sigma_{3}$ are the numbers such that: <br>
$\sigma_{1} = \sqrt{\lambda_{1}}, \sigma_{2} = \sqrt{\lambda_{2}}$

$\Rightarrow$ <b>Here, singular values of $A$ are $\sigma_{1} = \sqrt{1} = 1$, and $\sigma_{2} = \sqrt{4} = 2$</b>

$\Rightarrow D = \begin{pmatrix}
\sigma_{1}&0\\

\end{pmatrix} = \begin{pmatrix}

\end{pmatrix}$

$\Rightarrow y = 0$ and $x$ can be any non-zero $s$. Thus $v = \begin{pmatrix}

\end{pmatrix}$

and normalized eigenvector $v_{1} = \frac{1}{\lVert v \rVert}v = \frac{1}{1}v = \begin{pmatrix}

\end{pmatrix}$


Eigenvector of $\lambda_{2}$

$\begin{pmatrix}

\end{pmatrix}$

$\Rightarrow \begin{pmatrix}

\end{pmatrix}$

$\Rightarrow x = 0$ and $y$ can be any non-zero $s$. Thus $v = \begin{pmatrix}

and normalized eigenvector $v_{2} = \frac{1}{\lVert v \rVert}v = \frac{1}{1}v = \begin{pmatrix}

Therefore $V = (v_{1}\quad v_{2})  =\begin{pmatrix}
1&0\\

\end{pmatrix}$

Next, <br>
using $\sigma_{1}u_{1} = Av_{1}$ and $\sigma_{2}u_{2} = Av_{2}$

we have, 

$u_{1} = \begin{pmatrix}

$\Rightarrow u_{2} = \frac{1}{2}\begin{pmatrix}

$\Rightarrow U = (u_{1}\quad u_{2}) = \begin{pmatrix}
1&0\\
0&1\\
\end{pmatrix}$

Since, $U = V = I$<br>
$A = UDV^T = IDI^T = A$

Prove proposition 7.31

<b> Proposition 7.31: Let $A$ be any matrix. Then the eigenvalues of $A^TA$ are positive or zero</b>

<b> Solution 2 </b> 

Let $\lambda_{1}, \lambda_{2}, \cdots, \lambda_{n}$ be the eigenvalues with eigenvectors $v_{1}, v_{2}, \cdots, v_{n}$.

Then, as we know $Av = \lambda v$, so for any randomly chosen  $v_{j}$

$A^TAv_{j} = \lambda_{j}v_{j}$

We also know that, <br>
if $A$ is any matrix then $A^TA$ is a symmetric matrix.

$\Rightarrow A^TA$ is orthogonally diagonalizable<br>
$\Rightarrow$ eigenvectors are orthonormal

Now taking norm square:<br>

$\lVert Av_{j} \rVert^2 = Av_{j}\cdot Av_{j}$

$\qquad = (Av_{j})^T Av_{j}\qquad$ (Since $(u\cdot v = u^T\cdot v))$
$\qquad = A^T V_{j}^T Av_{j}$<br>
$\qquad = v_{j}^T A^TAv_{j}$<br>
$\qquad = v_{j}^T \lambda_{j}v_{j}$<br>
$\qquad = \lambda_{j} (v_{j}\cdot v_{j}) = \lambda_{j}$<br>

Now, we have $\lambda_{j} = \lVert Av_{j} \rVert^2$ i.e., $\ge 0$

$\Rightarrow \lambda_{j} \ge 0$<br>

Hence, all eigenvalues of $A^TA$ are either positive values or zeroes.

$A  = UDV^T$

$\Rightarrow AA^T = UDV^T (UDV^T)^T$<br>

$\qquad = (UDV^T) (V^T)^T U^TD^T$

$\qquad = U (DD^T)U^T$

$\Rightarrow U^T (AA^T) U = D^{'}$

$\Rightarrow U$ diagonalizes $AA^T$ and the columns of $A$ are eigenvectors of $AA^T$

<b> Question 7</b> 

Prve that singular values of $A$ and $A^T$ are identical

<b> Solution 7 </b>

The singular values of a matrix $A$ are given by the square roots of the eigenvalues of $A^TA$: $\lambda_{1}, \lambda_{2},\cdots, \lambda_{n}$

Similarly, singular values of a matrix $A^T$ are given by the sqaure roots of the eigenvalues of $(A^T)^TA^T$: $\delta_{1}, \delta_{2}, \cdots, \delta_{n}$

We know that, <br>

The eigenvalues of the transposed matrix, $A^T$, are exactly the eigenvlues of $A$.

$\Rightarrow$ eigenvalues of transposed matrix $(A^T)^TA^T = $eigenvalues of $A^TA$

$\Rightarrow \delta_{1} = \lambda_{1}, \delta_{2} = \lambda_{2}, \cdots, \delta_{n} = \lambda_{n}$

$\Rightarrow \sqrt{\delta_{1}} = \sqrt{\lambda_{1}}, \sqrt{\delta_{2}} = \sqrt{\lambda_{2}}, \cdots, \sqrt{\delta_{n}} = \sqrt{\lambda_{n}}$

Hence, proved singular value of $A^T$ and $A$ are equal
