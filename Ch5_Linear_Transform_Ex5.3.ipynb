(i)ker$(T)\qquad$(ii)nullity$(T)\qquad$(ii)range$(T)\qquad$(iv) rank$(T)$

for the following linear transformation without using the dimension theorem:

(a)$T: \mathbb{R}^2 \to \mathbb{R}^2$ given by $T(v) = Av$ where $A = \begin{pmatrix}
1&2\\
1&2\\
\end{pmatrix}$

(b)$T: \mathbb{R}^3 \to \mathbb{R}^3$ given by $T(v) = Av$ where $A = \begin{pmatrix}

(d)$T: P_3 \to P_3$ given by $T(p) = xp'$

(e)$T: P_3 \to P_2$ given by $T(p) = p'$

(f)$T: P_3 \to \mathbb{R}$ given by $T(p) = \int_0^1P(x) dx$

(g)$T: M_{22} \to P_1$ given by $T\begin{pmatrix}\begin{pmatrix}
a&b\\
c&d\\
\end{pmatrix}\end{pmatrix} = (a+c)x + (b+d)$ 

<b> Solution (a) </b>

<b>(i)</b><br>
Let $v = \begin{pmatrix}
x\\
y\\
\end{pmatrix}$. For ker$(T)$, we need to find $x$ & $y$ such that $Tv = Av = O$

$\Rightarrow \begin{pmatrix}

\end{pmatrix}\begin{pmatrix}
x\\
y\\
\end{pmatrix} = O$

$\Rightarrow x + 2y = 0$<br>
and, $x + 2y = 0$

If $y = r$, then $x = -2r$ 

$\Rightarrow v = \begin{pmatrix}
-2r\\
r\\
\end{pmatrix}$ or $r\begin{pmatrix}
-2\\
1\\
\end{pmatrix}$

Hence, ker$(T) = \{ r\begin{pmatrix}
-2\\
1\\
\end{pmatrix} | r \in \mathbb{R}^n \}$


<b>(ii)</b> 

$nullity(T) = dim(T)$.<br>
where, $dim(T)$ is number of vectors in the basis for $ker(T)$.

Here, basis = $\{\begin{pmatrix}
-2\\
1\\
\end{pmatrix}\}$

Therefore, $dim(ker(T)) =1$<br>
$\Rightarrow nullity(T) = 1$


<b> (iii)</b>

x + 2y\\
\end{pmatrix}$

Thus, $range(T) = \{ \begin{pmatrix}
a\\
a\\
\end{pmatrix} | a \in \mathbb{R} \}$ for both entries being same

<b>(iv)</b>

$rank(T) = dim(range(T))$

Dimension of range is the number of vectors in the basis for $range(T)$

here, basis $= \left[\begin{pmatrix}
1\\
1\\
\end{pmatrix}\right]$ and number of vectors is 1.

Thus, $dim(range(T) = 1$

$\Rightarrow rank(T) = 1$
