<b>Question 1</b>

Determine the eigenvalues and eigenvectors of $A = \begin{pmatrix}
5&2\\
-2&1\\
\end{pmatrix}$, and plot the eigenspace $E_{\lambda}$, and write down a set of basis vectors for $E_{\lambda}$

<b>Solution</b>

Solving $det(A - \lambda I) = 0$ to get eigenvalues

$det\left[\begin{pmatrix}
5&2\\
-2&1\\
\end{pmatrix} - \lambda\begin{pmatrix}
1&0\\
0&1\\
\end{pmatrix}\right] = 0$

$\Rightarrow \begin{pmatrix}
5-\lambda&2\\
-2&1-\lambda\\
\end{pmatrix} = 0$<br>

$\Rightarrow (5-\lambda)(1-\lambda) - 2(-2) = 0$<br>
$\Rightarrow 5- \lambda - 5\lambda  + \lambda^2 + 4 = 0$<br>
$\Rightarrow \lambda^2 - 6\lambda + 9 = 0$<br>
$\Rightarrow (\lambda - 3)^2 = 0$

$\Rightarrow \lambda_{1,2} = 3$

For eigenvector, solve $(A - \lambda I)u = 0$

$\left(\begin{pmatrix}
5&2\\
-2&1\\
\end{pmatrix} -3\begin{pmatrix}
1&0\\
0&1\\
\end{pmatrix}\right)\begin{pmatrix}
x\\
y\\
\end{pmatrix} = \begin{pmatrix}
0\\
0\\
\end{pmatrix}$

$\Rightarrow \begin{pmatrix}
5-3&2\\
-2&1-3\\
\end{pmatrix} \begin{pmatrix}
x\\
y\\
\end{pmatrix} = \begin{pmatrix}
0\\
0\\
\end{pmatrix}$

$\Rightarrow \begin{pmatrix}
2&2\\
-2&-2\\
\end{pmatrix} \begin{pmatrix}
x\\
y\\
\end{pmatrix} = \begin{pmatrix}
0\\
0\\
\end{pmatrix}$

$\Rightarrow 2x + 2y = 0$<br>
$\Rightarrow -2x -2y = 0$<br>

$y = -x$. Therefore, eigenvector, $u = \begin{pmatrix}
1\\
-1\\
\end{pmatrix}$ and similarly, basis vector for $E_{\lambda} = \begin{pmatrix}
1\\
-1\\
\end{pmatrix}$

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from scipy.linalg import eig
```

```python
A = np.array([[5, 2],[-2, 1]])

# Find eigenvalues, eigenvectors, and extracting eigenvectors off it
eigenvalues, eigenvectors = eig(A)
v1 = eigenvectors[:, 0]
v2 = eigenvectors[:, 1]

# Create a list of points along the eigenspaces
t = np.linspace(-15, 15, 100)

# Eigenspace 1: span of the first eigenvector
eigenspace1 = np.outer(v1, t)

# Eigenspace 2: span of the second eigenvector
eigenspace2 = np.outer(v2, t)

plt.figure(figsize=(10, 8))
plt.xlim(-8, 8)
plt.ylim(-8, 8)
plt.grid()


# Plot the eigenspaces
plt.plot(eigenspace1[0, :], eigenspace1[1, :], label="Eigenspace 1", color="blue")
plt.plot(eigenspace2[0, :], eigenspace2[1, :], label="Eigenspace 2", color="green")

plt.axhline(0, color='black', linewidth=1)
plt.axvline(0, color='black', linewidth=1)
for x_val in range(-8, 9):
    plt.plot(x_val, 0, 'k+', markersize=5)
    plt.text(x_val, -0.5, f'{x_val}')

for y_val in range(-8, 9):
    plt.plot(0, y_val, 'k+', markersize=5)
    plt.text(-0.5, y_val, f'{y_val}')

plt.title('Eigenspaces for Matrix A')
plt.legend(loc="upper right")
plt.show()
```

<b>Question 3</b>

Let $A$ be a $2$ by $2$ matrix. Show that the characteristic polynomial $p(\lambda)$ is given by :

$p(\lambda) = \lambda^{2} - tr(A)\lambda + det(A)$

where $tr(A)$ is the trace of the matrix $A$

<b> Solution</b> 

If our $2 \times 2$ matrix is $A = \begin{pmatrix}
a&b\\
c&d\\
\end{pmatrix}$, then<br>

$det(A - \lambda I)$<br>
$\quad = det\left(\begin{pmatrix}
a&b\\
c&d\\
\end{pmatrix} - \lambda \begin{pmatrix}
1&0\\
0&1\\
\end{pmatrix}\right)$

$\quad = det\begin{pmatrix}
a-\lambda&b\\
c&d-\lambda\\
\end{pmatrix}$

$\quad = (a - \lambda)(d - \lambda) - bc$<br>
$\quad = (\lambda -a)( \lambda -d) - bc$<br>
$\quad = \lambda^2 - a\lambda - d\lambda + ad - bc$<br>

$\quad = \lambda^2 - (a+b)\lambda + (ad - bc)$<br>

We know that,<br>
tr(A) here would be $(a+b)$ and $det$ would be $ad - bc$

Therefore, characteristic polynomila $p(\lambda)$ or $det(A - \lambda I)$<br>

$\quad = \lambda^2 - tr(A)\lambda + det(A)$<br>

Hence, proved!

<b>Question 4</b>

Let $A$ be a 2 by 2 matrix with $tr(A) = 2a$ and $det(A) = a^2$ where $a$ is a real number. Show that this matrix has an eigenvalue of $a$ with multiplicity of 2.

<b> Solution </b> 

If our $2 \times 2$ matrix is $A = \begin{pmatrix}
a&b\\
c&d\\
\end{pmatrix}$, then<br>

$det(A - \lambda I)$<br>
$\quad = det\left(\begin{pmatrix}
a&b\\
c&d\\
\end{pmatrix} - \lambda \begin{pmatrix}
1&0\\
0&1\\
\end{pmatrix}\right)$

$\quad = det\begin{pmatrix}
a-\lambda&b\\
c&d-\lambda\\
\end{pmatrix}$

$\quad = (a - \lambda)(d - \lambda) - bc$<br>
$\quad = (\lambda -a)( \lambda -d) - bc$<br>
$\quad = \lambda^2 - a\lambda - d\lambda + ad - bc$<br>

$\quad = \lambda^2 - (a+b)\lambda + (ad - bc)$<br>

We know that,<br>
tr(A) here would be $(a+b)$ and $det$ would be $ad - bc$

Thus, characteristic polynomial $p(\lambda)$ or $det(A - \lambda I)$<br>

$\quad = \lambda^2 - tr(A)\lambda + det(A)$<br>

We have been given $tr(A) = 2a$ and $det(A) = a^2$, therefore, 

$\quad \lambda^2 - tr(A)\lambda + det(A)$<br>
$\quad = \lambda^2 - 2a\lambda + a^2$<br>
$\quad = (\lambda^ - a)^2$

$\Rightarrow \lambda_{1,2} = a$

Therefore, eigenvalue $a$ with multiplicity of $2$

<b>Question 5</b>

Determine the eigenvalues and eigenvectors of $A = \begin{pmatrix}
1&0&0\\
-3&1&0\\
7&9&1\\
\end{pmatrix}$. By using appropriate software, plot the eigenspace $E_{\lambda}$ and write down a set of basis vectors for this eigenspace.

<b> Solution</b>

We know that, <br>
If a $n \times n$ matrix $A$ is diagonal or triangular matrix then the eigenvalues of $A$ are the entries along the leading diagonal

Here, $A = \begin{pmatrix}
1&0&0\\
-3&1&0\\
7&9&1\\
\end{pmatrix}$ is a lower triangular matrix, therefore, its eigenvalues are all 1. ie., $\lambda_{1, 2, 3} = 1$

Now, to find eigenvector we solve $(A - \lambda I)u = 0$

Here, <br>

$(A - \lambda I)u = 0$

$\Rightarrow \begin{pmatrix}
1-1&0&0\\
-3&1-1&0\\
7&9&1-1\\
\end{pmatrix} \cdot\begin{pmatrix}
x\\
y\\
z\\
\end{pmatrix} = \begin{pmatrix}
0\\
0\\
0\\
\end{pmatrix}$

$\Rightarrow \begin{pmatrix}
0&0&0\\
-3&0&0\\
7&9&0\\
\end{pmatrix} \cdot\begin{pmatrix}
x\\
y\\
z\\
\end{pmatrix} = \begin{pmatrix}
0\\
0\\
0\\
\end{pmatrix}$

$\Rightarrow -3x = 0$<br>
and, $7x +9y = 0$

$\Rightarrow x = 0$ and $y = 0$

Then, $z$ can be any non-zero value $z = s$, and we can write eigenvector $u$ as $u = s\begin{pmatrix}
0\\
0\\
1\\
\end{pmatrix}$. Similarly, basis vector for $E_{1} = \begin{pmatrix}
0\\
0\\
1\\
\end{pmatrix}$

```python
A = np.array([[1, 0, 0],
              [-3, 1, 0],
              [7, 9, 1]])

eigenvalues, eigenvectors = eig(A)
eigenvectors = eigenvectors / np.linalg.norm(eigenvectors, axis=0)
fig = plt.figure(figsize=(6,6))
ax = fig.add_subplot(111, projection='3d')

origin = [0, 0, 0]
for i in range(len(eigenvalues)):
    ax.quiver(*origin, *eigenvectors[:, i], color='r')

ax.set_xlim([-1, 1])
ax.set_ylim([-1, 1])
ax.set_zlim([-1, 1])
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')
ax.view_init(elev=20, azim=40)  # Adjust the view for better visualization

plt.show()
```
