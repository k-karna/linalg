<b> Question 1</b> <br>
Let $u$ and $v$ be vectors in $\mathbb{R}^2$. For the following $u$ and $v$ determine the angle between the vectors and label this angle and the vectors.

(a) $u = \begin{pmatrix}
1\\
1\\
\end{pmatrix}$,$v = \begin{pmatrix}
0\\
1\\
\end{pmatrix}\qquad$
(b)$u = \begin{pmatrix}
1\\
0\\
\end{pmatrix}$,$v = \begin{pmatrix}
0\\
1\\
\end{pmatrix}\qquad $
(c)$u = \begin{pmatrix}
-2\\
3\\
\end{pmatrix}$, $v = \begin{pmatrix}
1/2\\
-1/2\\
\end{pmatrix}$

<b> Solution </b> <br> 
<b>(a)</b> <br> $cos(\theta) = \frac{u \cdot v}{\lVert u \rVert \cdot \lVert v \rVert}$

$u \cdot v = \begin{pmatrix}
1\\
1\\
\end{pmatrix} \cdot \begin{pmatrix}
0\\
1\\
\end{pmatrix} = \sqrt{1(0) + 1(1)} = 1$ 

$\lVert u \rVert \cdot \lVert v \rVert = \sqrt{1^2 + 2^2 \cdot} \sqrt{0^2 + 1^2} = \sqrt{2}$

$ cos(\theta) = \frac{1}{\sqrt2}$<br>
$ \Rightarrow \theta = 45^o$

<b>(b)</b> <br> $cos(\theta) = \frac{u \cdot v}{\lVert u \rVert \cdot \lVert v \rVert}$ 

$u \cdot v = \begin{pmatrix}
1\\
0\\
\end{pmatrix} \cdot \begin{pmatrix}
0\\
1\\
\end{pmatrix} = \sqrt{1(0) + 0(1)} = 0$

Since, $u \cdot v = 0$ <br>
$\Rightarrow \theta = 90^o$

<b>(c)</b> <br>
$cos(\theta) = \frac{u \cdot v}{\lVert u \rVert \cdot \lVert v \rVert}$ <br>
$u \cdot v = \begin{pmatrix}
-2\\
3\\
\end{pmatrix} \cdot \begin{pmatrix}
1/2\\
-1/2\\
\end{pmatrix} = \sqrt{-2(1/2) + 3(-1/2)} = -5/2$


$\lVert u \rVert \cdot \lVert v \rVert = \sqrt{(-2)^2 + 3^2} \cdot \sqrt{ (1/2)^2 + (-1/2)^2} = \sqrt{13} \cdot \sqrt{\frac{1}{2}}$

$\Rightarrow cos(\theta) = \frac{-5/2}{\sqrt{13} \cdot \sqrt{\frac{1}{2}}} = -0.9805$ <br>
$\Rightarrow \theta = cos^{-1}(-0.98)$

<b>Question 10</b><br>

Prove that the angle between a non-zero vector $u$ and $-u$ in any n-space, $\mathbb{R}^n$, is $\pi$ radians or 180 degree

<b> Solution </b> <br> 

We know, <br>

$cos(\theta) = \frac{u \cdot v}{\lVert u \rVert \cdot \lVert v \rVert}$

Here, $u = u$ and $v = -u$

Therefore, $u \cdot v = u \cdot (-u)$ <br>
$= u \cdot (-1)u$<br>
$= (-1)(u \cdot u)$<br>
$= (-1)\lVert u \rVert^2$ as$\qquad  \sqrt{u \cdot u} = \lVert u \rVert$

Next, $\lVert v \rVert$ <br>

$= \lVert -u \rVert$

we know that, $\lVert ku \rVert = \lvert k \rvert \lVert u \rVert$

Therefore, $\lVert -u \rVert = \lVert (-1)u \rVert$<br>
$\qquad\qquad\qquad =  \lvert -1 \rvert \lVert u \rVert$<br>
$\Rightarrow \quad \qquad \lVert v \rVert = 1\lVert u \rVert = \lVert u \rVert$

Now, <br>

$cos(\theta) = \frac{u \cdot v}{\lVert u \rVert \cdot \lVert v \rVert}$

$\Rightarrow \qquad cos(\theta) = \frac{(-1)\lVert u \rVert^2}{\lVert u \rVert \cdot \lVert u \rVert}$

$\Rightarrow \qquad cos(\theta) = \frac{(-1)\lVert u \rVert^2}{\lVert u \rVert^2} = -1$

$\Rightarrow \qquad \theta = cos^{-1}(-1) = 180^o$
