<b> Question 1</b><br>

Determine whether the following vectors span $\mathbb{R}^2$: 

(a) $e_1 = \begin{pmatrix}
1\\
0\\
\end{pmatrix}, e_2 = \begin{pmatrix}
0\\
1\\
\end{pmatrix}\qquad$ (b) $u = \begin{pmatrix}
1\\
1\\
\end{pmatrix}, v = \begin{pmatrix}
-1\\
1\\
\end{pmatrix}$ <br>(c) $u = \begin{pmatrix}
2\\
2\\
\end{pmatrix}, v = \begin{pmatrix}
-1\\
-1\\
\end{pmatrix}\qquad$(d) $u = \begin{pmatrix}
1\\
2\\
\end{pmatrix}, v = \begin{pmatrix}
-1\\
10\\
\end{pmatrix}\qquad$

<b> Solution</b> <br>

<b>(a)</b> Lets take an arbitary vector $w = \begin{pmatrix}
a\\
b\\
\end{pmatrix}$ and scalar $k$ and $c$ to make linear combination

$ke_1 + ce_2 = w$

$\Rightarrow k\begin{pmatrix}
1\\
0\\
\end{pmatrix}$ + $c\begin{pmatrix}
0\\
1\\
\end{pmatrix} =\begin{pmatrix}
a\\
b\\
\end{pmatrix}$

$\Rightarrow \begin{pmatrix}
k\\
0\\
\end{pmatrix}$ + $\begin{pmatrix}
0\\
c\\
\end{pmatrix} =\begin{pmatrix}
a\\
b\\
\end{pmatrix}$


$\Rightarrow \begin{pmatrix}
k\\
c\\
\end{pmatrix} = \begin{pmatrix}
a\\
b\\
\end{pmatrix}$

Here, $k=a$, $c=b$, which makes $w = ku + cv$. Therefore, $e_1$ and $e_2$ span $\mathbb{R}^2$

<b>(b)</b> Again, taking an arbitary vector $w = \begin{pmatrix}
a\\
b\\
\end{pmatrix}$ which makes $w = ku + cv$<br>

$\Rightarrow k\begin{pmatrix}
1\\
1\\
\end{pmatrix} + c\begin{pmatrix}
-1\\
1\\
\end{pmatrix} = \begin{pmatrix}
a\\
b\\
\end{pmatrix}$

$\Rightarrow \begin{pmatrix}
k-c\\
k+c\\
\end{pmatrix}$

This gives us: 
$k - c = a \dots eq(1)$<br>
$k + c = b \dots eq(2)$<br>

From both eq, we get $2k = a+b$ <br>
$\Rightarrow k = \frac{a+b}{2}$ Therefore, $c = \frac{b-a}{2}$ and span $\mathbb{R}^2$

<b>Solution (c)</b><br>

Let $w = \begin{pmatrix}
a\\
b\\
\end{pmatrix}$ satisfying $ku + cv = w$

Then, $ku + cv = k\begin{pmatrix}
2\\
2\\
\end{pmatrix} + c\begin{pmatrix}
-1\\
-1\\
\end{pmatrix} = \begin{pmatrix}
a\\
b\\
\end{pmatrix}$

$\Rightarrow \begin{pmatrix}
2k -c\\
2k -c\\
\end{pmatrix} = \begin{pmatrix}
a\\
b\\
\end{pmatrix}$

This gives us:<br>
2k -c = 0<br>
2k -c = 0<br>
$0 = a - b$ or $a = b$<br>

$\Rightarrow$ given vector $u$ and $v$ span $\begin{pmatrix}
a\\
a\\
\end{pmatrix}$ and do not span $\mathbb{R}^2$

<b>Solution (d)</b><br>

Let $w = \begin{pmatrix}
a\\
b\\
\end{pmatrix}$ satisfying $ku + cv = w$

Then, $ku + cv = k\begin{pmatrix}
1\\
2\\
\end{pmatrix} + c\begin{pmatrix}
-1\\
10\\
\end{pmatrix} = \begin{pmatrix}
a\\
b\\
\end{pmatrix}$

$\Rightarrow \begin{pmatrix}
k -c\\
2k +10c\\
\end{pmatrix} = \begin{pmatrix}
a\\
b\\
\end{pmatrix}$

This gives Augmented Matrix as :
$\left(\begin{array}{cc|c}
1&-1&a\\
2&10&b\\
\end{array}
\right)$

After Gaussian elimination it is :
$\left(\begin{array}{cc|c}
1&-1&a\\
0&12&b-2a\\
\end{array}
\right)$

$\Rightarrow 12c = b -2a$<br>
$\Rightarrow c = \frac{b -2a}{12}$<br>

Then, from the top row, we can deduce $k = a +c$<br>

$\Rightarrow k = a + \frac{b - 2a}{12}$ or $k = \frac{b + 10a}{12}$

Here, with both $k$ and $c$ scalar, given vector $u$ and $v$ span $\mathbb{R}^2$

<b> Question 3</b><br>


Determine whether the following vectors form a basis for $\mathbb{R}^2$: 

(a) $u = \begin{pmatrix}
1&2\\
\end{pmatrix}^T, v = \begin{pmatrix}
0&1\\
\end{pmatrix}^T\qquad$

(b) $u = \begin{pmatrix}
-2&-4\\
\end{pmatrix}^T, v = \begin{pmatrix}
1&2\\
\end{pmatrix}^T\$

(c) $u = \begin{pmatrix}
4&1\\
\end{pmatrix}^T, v = \begin{pmatrix}
1&3\\
\end{pmatrix}^T\qquad$

(d) $u = \begin{pmatrix}
3&5\\
\end{pmatrix}^T, v = \begin{pmatrix}
2&3\\
\end{pmatrix}^T\qquad$

<b> Solution: <br>
    Key thing</b> There are two conditions to be satisfied to establish if $n$ vectors form a basis. <br>
    1. The vectors span $\mathbb{R}^n$<br>
    2. The vectors are linearly independent

However, if we have $n$ vectors in $\mathbb{R}^n$, then we only need to <b>check one of the conditions, either independence or span</b>

<i>This section is about span. We will check span</i>

<b>(a)</b> Here, we have, $u = \begin{pmatrix}
1\\
2\\
\end{pmatrix}$ and $v = \begin{pmatrix}
0\\
1\\
\end{pmatrix}$

To check span, lets take $k$ and $c$ such that $ku + cv = w$ where $w$ is an arbitary vector $\begin{pmatrix}
a\\
b\\
\end{pmatrix}$

Now<br>
$k\begin{pmatrix}
1\\
2\\
\end{pmatrix} + c\begin{pmatrix}
0\\
1\\
\end{pmatrix} =\begin{pmatrix}
a\\
b\\
\end{pmatrix}$

$\Rightarrow \begin{pmatrix}
k\\
2k+c\\
\end{pmatrix} = \begin{pmatrix}
a\\
b\\
\end{pmatrix}$

$\Rightarrow k =a$ and $c = b - 2k \quad\text{or}\quad c = b-2a$

Therefore, $u$ and $v$ span $\mathbb{R}^2$<br>
Hence, $u$ and $v$ <b>form basis</b>


<b>(b)</b> Here, we have, $u = \begin{pmatrix}
-2\\
-4\\
\end{pmatrix}$ and $v = \begin{pmatrix}
1\\
2\\
\end{pmatrix}$

To check span, lets take $k$ and $c$ such that $ku + cv = w$ where $w$ is an arbitary vector $\begin{pmatrix}
a\\
b\\
\end{pmatrix}$

Now<br>
$k\begin{pmatrix}
-2\\
-4\\
\end{pmatrix} + c\begin{pmatrix}
1\\
2\\
\end{pmatrix} =\begin{pmatrix}
a\\
b\\
\end{pmatrix}$

$\Rightarrow 
\begin{pmatrix}
-2k+ c\\
-4k + 2c\\
\end{pmatrix} =\begin{pmatrix}
a\\
b\\
\end{pmatrix}$

This gives Augmented Matrix as :
$\left(\begin{array}{cc|c}
-2&1&a\\
-4&2&b\\
\end{array}
\right)$

After Gaussian elimination it is :
$\left(\begin{array}{cc|c}
-2&1&a\\
0&0&b\\
\end{array}
\right)$

Here, $b = 0$ <br>
Therefore, $u$ and $v$ do not span $\mathbb{R}^2$<br>
Hence, $u$ and $v$ <b>do not form basis</b>


<b>(c)</b><br>

Here, we have, $u = \begin{pmatrix}
4\\
1\\
\end{pmatrix}$ and $v = \begin{pmatrix}
1\\

\end{pmatrix}$

To check span, lets take $k$ and $c$ such that $ku + cv = w$ where $w$ is an arbitary vector $\begin{pmatrix}

Now<br>

This gives us:


$4k + c = a$<br>
$k + 3c = b$<br>

Solving above equations, we get $k = \frac{3a - b}{11}$ and $c = \frac{-(a+b)}{11}$

Therefore, $u$ and $v$ span $\mathbb{R}^2$<br>
Hence, $u$ and $v$ <b>form basis</b>



<b>(d)</b><br>

Here, we have, $u = \begin{pmatrix}
3\\
5\\
\end{pmatrix}$ and $v = \begin{pmatrix}
2\\

\end{pmatrix}$

To check span, lets take $k$ and $c$ such that $ku + cv = w$ where $w$ is an arbitary vector $\begin{pmatrix}

\end{pmatrix} + c\begin{pmatrix}

\end{pmatrix} =\begin{pmatrix}
a\\
b\\
\end{pmatrix}$

This gives us:


$3k + 2c = a$<br>
$5k + 3c = b$<br>

Solving above equations, we get $k = 2b - 3a$ and $c = 2a - b$

Therefore, $u$ and $v$ span $\mathbb{R}^2$<br>
Hence, $u$ and $v$ <b>form basis</b>

k_2\\
k_3\\
\end{pmatrix}$, we need to find values of $x$ such that $Ax = b$ ie.,<br>

k_2\\

$\left(\begin{array}{ccc|c}

\right)$

$\Rightarrow\left(\begin{array}{ccc|c}

$3k_2 = 1$<br>

<b> Question 13</b><br>


Prove proposition 2.27<br>
"Any $n$ linearly independent vectors in $\mathbb{R}^n$" form a basis for $\mathbb{R}^n$


<b>Solution</b>

We know that there is a proposition that states


<b>"If we have $v_1, v_2, v_3,\dots v_m$ as different vectors in $\mathbb{R}^n$, where $n \lt m$, then the vectors $v_1, v_2, v_3, \dots v_m$ are linearly dependent"</b>

Now, suppose if we have a set of $n$ independent vectors, S i.e, $S = \{v_1, v_2, v_3, \dots v_n\}$ that does not span $\mathbb{R}^n$.

Then, there must be another $u$ that is not a linear combination of $S$.<br>

Lets call this new set as $T$ which consists of now $n+1$ vectors in $\mathbb{R}^n$ and written as $T = \{v_1, v_2, v_3, \dots v_n, u\}$

From earlier proposition, this set $T = \{v_1, v_2, v_3, \dots v_n, u\}$ must be linearly dependent

$\Rightarrow c_1v_1 + c_2v_2 + \dots c_nv_n + c_{n+1}u = 0$ where all c's are not zero.

$\Rightarrow u = (-c_1/c_{n+1})v_1 + (-c_2/c_{n+1})v_2 + ... + (-c_n/c_{n+1})v_n$

We can see $u$ can be written as linear combination of $v_1, v_2, v_3, \dots v_n$, the set of vectors $S$.

Therefore, $u$ is in the span of $S$

This contradiction our prior assumption that $u$ cannot be written as linear combination of $S$ and $S$, as set of linearly independent vectors does not span $\mathbb{R}^n$

<b>Hence, $S = \{v_1, v_2, v_3, \dots v_n\}$  is a set of $n$ linearly independent vectors in $\mathbb{R}^n$ so it a basis for $\mathbb{R}^n$</b>

<b> Question 14</b><br>

Prove proposition 2.28<br>
"Any $n$ vectors which span $\mathbb{R}^n$ form a basis for $\mathbb{R}^n$"

<b>Solution</b>

Let $S = \{v_1, v_2, v_3, \dots v_n\}$ be set of $n$ vectors which span $\mathbb{R}^n$

As we know from Question 13, <b>"Any $n$ linearly independent vectors in $\mathbb{R}^n$ form a basis for $\mathbb{R}^n$</b>

We need to prove $S = \{v_1, v_2, v_3, \dots v_n\}$ are linearly independent to be a basis for  $\mathbb{R}^n$

Suppose the vectors in $S$ are linearly dependent.<br>
Then one of the vectors can be written as liner combination of others.

If that vector is $v_n$ and the remaining vectors $T = {v_1, v_2, \dots, v_n}$ still spans $\mathbb{R}^n$

If we keep repeating this process, the very last set of vectors would still span $\mathbb{R}^n$ and linearly independent to form a basis for $\mathbb{R}^n$

This contradicts, one proposition <b>"Every basis of $\mathbb{R}^n$ contains exactly $n$ vectors". Therefore our supposition $S = \{v_1, v_2, v_3, \dots v_n\}$ to be linearly dependent must be wrong.
    
Hence, any $n$ vectors which span $\mathbb{R}^n$ form a basis for $\mathbb{R}^n$
